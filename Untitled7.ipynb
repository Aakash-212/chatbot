```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

data ={
    "text":["hello","hii","bye","good bye","where are you","what is your name"],
    "intent":["greeting","greeting","bye","bye","name","name"]
}
df = pd.DataFrame(data)
X = df["text"]
Y = df["intent"]

cv = CountVectorizer()
X_vector = cv.fit_transform(X)

model = MultinomialNB()
model.fit(X_vector,Y)

while True:
  user_input = input("you: ")

  if user_input.lower() =="exit":
    break
  user_vector = cv.transform([user_input])
  prediction = model.predict(user_vector)

  if prediction[0] == "greeting":
      print("Bot: Hello!")
  elif prediction[0] == "bye":
      print("Bot: Goodbye!")
  elif prediction[0] == "name":
      print("Bot: I am your chatbot")
```
